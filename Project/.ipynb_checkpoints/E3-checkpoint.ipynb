## Q1. Characterize Model

Consider a stylized, static model of the electricity wholesale market where a portfolio of different dispatchable generators (indexed by $i$) competes to serve a fixed load $L$ MWh. Let $E_i$ denote the generated electricity by generator $i$ measured in MWh and $q_i$ the installed hourly generating capacity in MW. The generators use a combination of primary energy carriers (indexed by $j$) and an aggregate of other inputs ($oth$). Let $\mu_j^i$ denote the input share of energy carrier $j$ for generator $i$ -- i.e. MWh input of energy carrier $j$ per MWh electricity output. Let $p_j^f$ denote the price of carrier $j$ measured in ‚Ç¨/MWh. The use of primary carriers ($j$) is tied to \CO2 emissions in fixed ratios.

### **Q1.1: Briefly explain the optimization problem in equations (1).**
$\begin{align}
    \min \ \sum_i& c_i E_i \tag{1a} \\
    \text{subject to }L&=\sum_i E_i \tag{1b} \\ 
    c_i &= c_i^{oth}+\sum_j\mu_j^i p_j^f, \qquad \forall i \tag{1c}\\ 
    E_i &\in[0, q_i], \qquad \forall i. \tag{1d}
\end{align}$

Answer: 
    The Objective Function (Eqn. 1a): Our overall optimisation problem is to minimise the total system cost of the energy generators i. As 1a) represents the total short run costs of producing 1 GJ of electricity, we are minimising the total cost c_i*E_i across these generators. 
    Equilibrium Constraint (Eqn. 1b): This minimisation is subject to the equilbrium constraint. Our overall fixed load L MWh is equal to the sum of production over all generators.
    Cost Structure (Eqn. 1c): This equation describes the actual cost we need to minimise, where for all generators, the costs are equal to the costs of other energy inputs (c^oth) i.e. labour, maintenance, materials etc. plus the price on fuel (p) multiplied by the GJ of fuel generator to produce 1 GJ of fuel (mu). This is then summed over all the energy carriers j for this reactor i. 
    Bound Constraint (Eqn. 1d): This constraint basically says that for all reactors we cannot produce -ve amounts of electricity E_i and our production cannot exceed the generator's technical capacity (q_i)

### **Q1.2:**

*Let $\phi_j$ denote the emission intensity of energy carrier $j$ measured in ton CO2/MWh. Assume that we have an estimate of the social cost of carbon of $p_{CO2}$ ‚Ç¨/ton CO2.*


*Briefly explain the problem in equations (2). What does the new parameter $v_i$ measure?*
\begin{align}
    \min \ \sum_i& \left(c_i+v_i\cdot p_{CO2}\right) E_i \tag{2a}\\
    \text{subject to }L&=\sum_i E_i \tag{2b} \\ 
    v_i &= \sum_j\mu_j^i\phi_j \tag{2c} \\
    E_i &\in[0, q_i], \qquad \forall i. \tag{2d}
\end{align}

Answer: As $\mu_j^i$ is the input share for the energy carrier j to the generator i and $\phi_j$ is the emission intensity of the specific carrier -> summing the product of the input share of each energy carrier and its associated CO‚ÇÇ emissions intensity across all energy carriers j used by generator
i for all of the energy carriers across the reactor i means that v_i is the total emissions per MWh of electricity generated by the generator i. 

Hence, this optimisation problem is similar to the one above, except we are minimising the CO‚ÇÇ emissions for E_i MWh of electricity across our reactors as well. Further, this aims to minimise the social cost of carbon by multiplying the cost of the estimate price ùëùùê∂ùëÇ2 by our emissions intensity. This is subject to the same bound and equilibrium constraints seen above

### **Q1.3**
*Derive the first order conditions for the problem in (2b). Give a brief interpretation.*

Lagrangian:<br>
$\mathcal{L}^B$ = $f(x) + \lambda_h(k_h - h(x))$ + $\lambda_g(k_g) - g(x)) + \bar{\theta^x}(\bar{x} - x) + \underline{\theta_x}(x - \bar{x})$ <br>

$\mathcal{L}^B$ = $-\sum_i (c_i+v_i*p_{CO2}) * E_i + \lambda(\sum_i (E_i) - L) - \bar{\theta^E}(q_i - E_i) + \underline{\theta^E}(E_i - 0)$<br>
**FOC's**:<br>
$‚àÇ\mathcal{L}/‚àÇE_i = -(c_i + v_i *p_{CO2}) + \lambda - \overline{\theta^E} + \underline{\theta_E} = 0$<br>
**Complementary Slackness Conditions:**<br>


**Interpretation** <br>
$\lambda$ is a shadow variable in the equilibrium constraint (2b). It measures the marginal system costs and hence the equilibrium price. If a cost for a specific generator $i$ ( $(c_i + v_i *p_(CO2)$ ) is less than $\lambda$, it will be operating at 100% capacity. If it is greater, it will operate at 0% capacity. If the generator cost = $\lambda$, then it will have an operational capacity of $E_i \in (0,q_i)$. But due to our equilibrium constraint, the $E_i$ for all generators $i$ must always sum to $L$

*Based on this small dataset, draw the merit-order curve and compute the solution for $E_i$ and $\lambda$.*

Note: I calculated these costs with the emissions costs (i.e. the Social MC)


````python
def calculate_cost(generator: str) -> float:
    """
    Calculates social cost for each generator"
    """
    c_i = db['OtherMC'][generator] + (db['FuelMix']['Coal'][generator] * db['FuelPrice']['Coal'] + db['FuelMix']['Natural gas'][generator] * db['FuelPrice']['Natural gas'])
    v_i = db['FuelMix']['Coal'][generator]*db['EmissionIntensity']['Coal']['CO2'] + db['FuelMix']['Natural gas'][generator]*db['EmissionIntensity']['Natural gas']['CO2']
    cost = c_i + (v_i * db['EmissionTax']['CO2'])
    return cost
````

````python
#Iterates through generators and places costs into list
generators = ['A', 'B', 'C']
cost_list = []
for generator in generators:
    cost_list.append(calculate_cost(generator))
capacity_list  = db['GeneratingCapacity']
````

````python
import matplotlib.pyplot as plt
import numpy as np
labels = ['A (Coal)', 'B (Natural Gas)', 'C (Wind)']

# Sort the capacity_list and cost_list based on the ascending order of cost_list
sorted_indices = np.argsort(cost_list)  # indices of sorted costs
sorted_capacities = [capacity_list[i] for i in sorted_indices]
sorted_costs = [cost_list[i] for i in sorted_indices]
sorted_labels = [labels[i] for i in sorted_indices]

# Generate x-axis for the step change graph
x_values = np.cumsum([0] + sorted_capacities)  # cumulative sum for the x-axis

# Create the step change graph using actual costs for y-values
plt.step(x_values, [0] + sorted_costs, where='post', color='blue', label='Generation')

# Annotate the graph with the labels below the step line
for i, label in enumerate(sorted_labels):
    plt.text(x_values[i] + (x_values[i + 1] - x_values[i]) / 2, sorted_costs[i] - 4, label, ha='center')

# Add a green vertical line at x = 100 and label it
plt.axvline(x=100, color='green', linestyle='--', label='Load')

# Extend the x-axis to finish after the plateau for the third step
plt.xlim(0, x_values[-1]-5)


# Set labels for axes
plt.xlabel('GJ')
plt.ylabel('‚Ç¨/GJ')

# Add both the blue step change and the green vertical line to the legend
plt.legend()
plt.title('Figure 1: Merit order curve with 3 plants and a fixed load')

# Show the graph
plt.show()
````

### **Q1.5. A Cap on Emissions**

Assume that, instead of having a price on CO2 emissions $(p_{CO2})$, we implement a cap on CO2 emissions as follows:
\begin{align}
    \sum_i v_i\cdot E_i \leq \overline{CO2}. \tag{4}
\end{align}
*Adjust the Lagrange in (3) appropriately and derive the first order conditions again.*

Hence:
$\begin{align}
    \min \ \sum_i& c_i E_i \tag{1a} \\
    \text{subject to }L&=\sum_i E_i \tag{1b} \\ 
    c_i &= c_i^{oth}+\sum_j\mu_j^i p_j^f, \qquad \forall i \tag{1c}\\ 
    \sum_i v_i\cdot E_i \leq \overline{CO2} \tag{1e}\\
    E_i &\in[0, q_i], \qquad \forall i. \tag{1f}
\end{align}$

We adjust the lagrangian to add this constraint<br>
$\theta^{CO2}(\overline{CO2} - \sum_i{v_i *E_i})$<br>
such that:
$\mathcal{L}^B$ = $-\sum_i c_i*E_i + \lambda(\sum_i (E_i) - L) +\theta^{CO2}(\overline{CO2} - \sum_i{v_i *E_i}) - \bar{\theta^E}(q_i - E_i) + \underline{\theta^E}(E_i - 0)$<br>
FOC:<br>
$‚àÇ\mathcal{L^B}/‚àÇE_i = -c_i+ +\theta^{CO2}(\overline{CO2} - v_i) + \lambda - \overline{\theta^E} + \underline{\theta_E} = 0$<br>
Complementary Slackness Conditions:<br>
$\theta^{CO2}(\overline{CO2} - \sum_i{v_i *E_i}) = 0, \overline{CO2} - \sum_i{v_i *E_i} \ge 0, \theta^{CO2} \ge 0$
If Emissions Cap is Binding:<br>
$\sum_i v_i*E_i = \overline{CO2}, \theta^{CO2} \gt 0$











### **Q1.6. Interpretation of shadow variable $\theta^{CO2}$**

In **Q1.4** we used a price on\CO2 emissions of 5‚Ç¨. This resulted in roughly 27.2 tonnes of CO2 emissions in optimum. Now, in the emissions cap scenario (**Q1.5**), let $\theta^{CO2}$ denote the shadow cost of the constraint in (4) and assume that we fix the cap at $\overline{CO2}=$25 tonnes instead. 


*Give an interpretation of what $\theta^{CO2}$ measures. Based on the small data set, what is the value in this model with $\overline{CO2}=$25?*

Answer (Interpretation): $\theta^{CO2}$ measures the shadow price or marginal cost of reducing emissions under the binding emissions cap constraint. It measures the increase in the objective function (total cost) per additional unit of of CO2 emissions allowed. When the cap is binding, $\theta^{CO2}$ indicates how much the cost would increase if the cap was relaxed by one tonne of CO2. 

Answer (Numerical Solution): For $\theta^{CO2}$, we can find this by solving for $E_i$ using the FOC:
    
$E_i = (c_i - \lambda + \overline{\theta^E} - \underline{\theta^E})/\theta^{CO2}*v_i$
<br>
Substitute into binding constraint:
<br>
$\sum_i v_i (c_i - \lambda + \overline{\theta^E} - \underline{\theta^E})/\theta^{CO2}*v_i) = \overline{CO2}$
<br>
Simplify and solve for $\theta^{CO2}$<br>
$\theta^{CO2} = \sum_i(c_i - \lambda + \overline{\theta^E} - \underline{\theta^E})/\overline{CO2}$
<br>
Under the binding emissions cap, it is likely that neither the upper or lower bound is active for most of the E_i's, hence our $\theta^E$'s (lagrange multipliers) can be zero.

## Q2. Numerical implementation

The algorithm in ```scipy.optimize.linprog``` solves what we will refer to as an "augmented form" of the linear program (as opposed to the "standard form" that does not feature explicit equality constraints):

$$\begin{align} 
    &\min_{x} \ c^T\cdot x \tag{5a}\\ 
    &A_{ub}\times x \leq b_{ub} \tag{5b}\\ 
    &A_{eq}\times x  = b_{eq} \tag{5c}\\ 
    &l\leq x\leq u, \tag{5d}
\end{align}$$
where: 
* $x$ is the vector of choice variables of length ($N$).
* $c, l, u$ are coefficient vectors of the same length ($N$).
* $b_{eq}, b_{ub}$ are coefficient vectors of lengths $N_{eq}, N_{ub}$, 
* and $A_{eq}, A_{ub}$ are coefficient matrices of sizes $(N_{eq}\times N)$ and $(N_{ub} \times N)$ respectively.

We will generally refer to the inequality constraints $l\leq x \leq u$ as *domain constraints*, whereas the constraints that combine multiple variables as *variational constraints* (1B-1C). 

At its core, the ```scipy.optimize.linprog``` takes inputs in the form of arrays (e.g. ```np.arrays```), which means that we have to be careful with the ordering of variables and constraints: The $n$'th element in $c$ represents the same element as the $n$'th elements in $l,u$ and the $n$'th column vectors in $A_{ub}, A_{eq}$. Thus, when solving a model with many different types of constraints and variables, the task of constructing suitable vectors / matrices becomes quite cumbersome. In the simple model from **Q1**, however, it is very simple.

### **Q2.2. Add the emission cap from (4) with $\overline{CO2} = 25$ to the model and solve.**

To add this emissions cap, we can rewrite:<br>
$A_{ub} = [v_1,...,v_i]$<br>
$b_{ub} = \overline{CO2}$

````python
#Hence
C02_max = 25
def calc_vi(generator):
    return db['FuelMix']['Coal'][generator]*db['EmissionIntensity']['Coal']['CO2'] + db['FuelMix']['Natural gas'][generator]*db['EmissionIntensity']['Natural gas']['CO2']
v_list = []
for generator in generators:
    v_list.append(calc_vi(generator))
A_ub = [[v_i for v_i in v_list]]
b_ub = [C02_max]
````

````python
optimize.linprog(c, A_eq = A_eq, b_eq = b_eq, A_ub = A_ub, b_ub = b_ub, bounds = np.vstack([l, u]).T)
````
